Dazu folgendes Beispiel:

```python
total = 14.40
vat_rate = 0.026
text = f"Included VAT ({(vat_rate * 100):.2f}%)"

vat = total * vat_rate
print(f"{text:<25}{vat:>7.2f} CHF")
```
Damit ein Text zusätzlich formatiert wird in einem f-String, wird nach der Variable ein ```:``` angegeben. Zum Beispiel ```{text:<25}``` fügt dem Text "Total inkl. Mwst (2.6%)" 25-len(text) Leerzeichen hinter an. Damit werden immer total 25 Zeichen ausgegeben. Umgekehrt fügt ```{text:>25}``` 25-len(text) Leerzeichen vor die Variable. ```<```orientiert den Text also linksbündig und ```>``` rechtsbündig. Damit lassen sich sehr leicht tabelarische Darstellungen verwirklichen. Zusätzlich kann man Zahlen und Daten formatieren. ```:>6.2f``` formatiert die Zahl ```vat``` mit Total 6 (```>6```) Zeichen (aufgefüllt mit Leerzeichen) nach rechts und 2 Nachkommastellen (```.2f```).

Weiter kann z.B. ein Datum formatiert werden.

```python
from datetime import datetime
now = datetime.now()
print(f"{now:%d.%m.%Y %H:%M}")
```
Wobei ```%d``` den Tag, ```%m``` den Monat und ```%Y``` das Jahr formatiert und ```%H``` die Stunden, ```%M``` die Minuten und ```%S``` die Sekunden. Wobei wir die Sekunden nicht brauchen und deshalb hier aus dem Format auslassen.
